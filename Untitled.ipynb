```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
tf.__version__
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   rotation_range = 40,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   fill_mode='nearest',
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory('dataset/training_set',
                                                 target_size = (150, 150),
                                                 batch_size = 32)
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255,
                                   rotation_range = 40,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   fill_mode='nearest',
                                   horizontal_flip = True)
test_set = test_datagen.flow_from_directory('dataset/test_set',
                                            target_size = (150, 150),
                                            batch_size = 32)
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[150, 150, 3]))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=120, activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=2, activation='softmax'))
```

```python
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['acc'])
```

```python
cnn.fit(x = training_set, validation_data = test_set, epochs = 25)
```

```python
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('dataset/single_prediction/mask2.jpg', target_size = (150, 150))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)
training_set.class_indices
print(result)
```

```python
import cv2
import numpy as np
```

```python
haar_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
results={0:'without mask',1:'mask'}
GR_dict={0:(0,0,255),1:(0,255,0)}
```

```python
rect_size = 4
cap = cv2.VideoCapture(0)
while True:
    (rval, im) = cap.read()
    im=cv2.flip(im,1,1) 
    
    rerect_size = cv2.resize(im, (im.shape[1] // rect_size, im.shape[0] // rect_size))
    faces = haar_data.detectMultiScale(rerect_size)
    for f in faces:
        (x, y, w, h) = [v * rect_size for v in f] 
        face_img = im[y:y+h, x:x+w]
        test_image = cv2.resize(face_img,(150,150))
        test_image = image.img_to_array(test_image)
        test_image = np.expand_dims(test_image, axis = 0)
        result = cnn.predict(test_image)
        print(result)
        label=np.argmax(result,axis=1)[0]
#         rerect_sized=cv2.resize(face_img,(150,150))
#         normalized=rerect_sized/255.0
#         reshaped=np.reshape(normalized,(1,150,150,3))
#         reshaped = np.vstack([reshaped])
#         result=cnn.predict(reshaped)
#         label=np.argmax(result,axis=1)[0]
#         print(label)
        
#         if result[0][0] > 0.5:
#             label = 0
#             prediction = "No-Mask"
#         else:
#             label = 1
#             prediction = "Mask"
    
      
        cv2.rectangle(im,(x,y),(x+w,y+h),GR_dict[label],2)
        cv2.rectangle(im,(x,y-40),(x+w,y),GR_dict[label],-1)
        cv2.putText(im, results[label], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
    cv2.imshow('LIVE',   im)
    key = cv2.waitKey(10)
    
    if key == 27: 
        break
cap.release()
cv2.destroyAllWindows()
```
